```python
# Load excel file
import openpyxl as xl
wb = xl.load_workbook(filename = 'maven_ski_shop_data.xlsx')
```

```python
# Create list without including the column title
info = wb['Orders_Info']
subtotals = []

for row in range(2, info.max_row + 1):
    subtotals.append(info[f'D{row}'].value)

subtotals
```

```python
# Calculate discount prices for customers that meet a certain criteria
discount_prices = []

for total in subtotals:
    if total > 500:
        discount_prices.append(total - (total*.2))
    else:
        discount_prices.append(total)

discount_prices
```

```python
# Create a dictionary of customer ID's and their subtotals
info = wb['Orders_Info']

cust_subtotals = []
cust_id = []

for row in range(2, info.max_row + 1):
    cust_subtotals.append(info[f'D{row}'].value)

for row in range(2, info.max_row + 1):
    cust_id.append(info[f'B{row}'].value)

cust_info = {}
for i, cust in enumerate(cust_id):
    cust_info[cust] = cust_subtotals[i]

cust_info
```

```python
cust_order_list = []

for row in range(2, info.max_row + 1):
    order_list = str(info[f'H{row}'].value).split(', ')
    cust_order_list.append(order_list)

cust_order_list
```

```python
updated_order_list = []
for cust_specs in cust_order_list:
    cust_set = []
    for cust_ind in cust_specs:
       cust_set.append(int(cust_ind))
    updated_order_list.append(cust_set)   

updated_order_list
```

```python
inventories = wb['Inventory_Levels']

no_inventory = []
for row in range(2, inventories.max_row + 1):
    if inventories[f'B{row}'].value == 0:
        no_inventory.append(int(inventories[f'A{row}'].value))
    
no_inventory
```

```python
inform_customer = []

for order_requests in updated_order_list:
    if any(single_number in order_requests for single_number in no_inventory):
        inform_customer.append('Alert customer')
    else:
        inform_customer.append('No need to contact')

inform_customer
```
